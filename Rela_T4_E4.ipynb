```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n y su gradiente
def f(x, y):
    return y - x**2

def df(x, y):
    dfx = -2*x
    dfy = np.ones_like(x)
    return dfx, dfy

# Rango de valores para x
x = np.linspace(-2, 2, 50)
y = x**2

# Calcula el gradiente en el rango de valores de x y y
dfx, dfy = df(x, y)

# Graficar la curva y el gradiente
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x, y) = y - x^2$', color='blue')
plt.quiver(x, y, dfx, dfy, scale=20, color='red', angles='xy', scale_units='xy', width=0.015, headwidth=5, headlength=7)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Curva y Gradiente df')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
