```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
```

a)

```python
# Definir la función a integrar
def integrand(t):
    return np.sqrt((t**2 + 2)**2 + (2*t)**2)

# Definir los límites de integración
a = 1  # Límite inferior
b = 5     # Límite superior

# Calcular la integral
result, error = integrate.quad(integrand, a, b)

# Imprimir el resultado
print("El valor de la integral es:", result)

# Graficar la función integranda
t_values = np.linspace(a, b, 100)
integrand_values = integrand(t_values)
plt.plot(t_values, integrand_values, label='Integranda')
plt.fill_between(t_values, integrand_values, color='skyblue', alpha=0.4, label='Área bajo la curva')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Función integranda y área bajo la curva')
plt.legend()
plt.grid(True)
plt.show()
```

b)

```python
def integrand(t):
    return np.sqrt((t**2 + 2)**2 - (2*t)**2)

result, error = integrate.quad(integrand, a, b)

print("El valor de la integral es:", result)

t_values = np.linspace(a, b, 100)
integrand_values = integrand(t_values)
plt.plot(t_values, integrand_values, label='Integranda')
plt.fill_between(t_values, integrand_values, color='skyblue', alpha=0.4, label='Área bajo la curva')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Función integranda y área bajo la curva')
plt.legend()
plt.grid(True)
plt.show()
```

Parametrización

```python
def parametrizacion(t):
    x1 = (1/3)*t**3 + 2*t + 2
    x2 = t**2 + 1
    return x1, x2

t_values = np.linspace(-5, 5, 1000)
x1_values, x2_values = parametrizacion(t_values)

plt.plot(x1_values, x2_values, label='Curva parametrizada', color='blue')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Parametrización de la curva R^2')
plt.legend()
plt.grid(True)
plt.show()
```
